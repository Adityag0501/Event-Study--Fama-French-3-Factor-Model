```python
import numpy as np
import matplotlib.pyplot as plt

import eventstudy as es
from eventstudy import excelExporter

from google.colab import files
```

## Single Event

```python
es.Single.import_returns('Final_Returns.csv')
es.Single.import_FamaFrench('Fama_French.csv')
```

```python
Ticker_List = ['JINDALSTEL', 'ULTRACEMCO', 'ADANIPORTS', 'BPCL', 'GAIL', 'PNB', 'SBIN', 'DLF', 'NESTLEIND']
for ticker in Ticker_List:
  event = es.Single.FamaFrench_3factor(
    security_ticker = ticker, 
    market_ticker='NSEI',
    event_date = np.datetime64('2021-02-01'),
    event_window = (-20,+20), 
    estimation_size = 200,
    buffer_size = 10
  )
  event.results().to_excel(f'{ticker}_single_results.xlsx')
  print('Printing 1st 5 rows in Event Results')
  print(event.results().head())
  event.plot(AR=True)
  print(f'{ticker} (FamaFrench-3 Factor)')
  plt.show() # use standard matplotlib function to display the plot
```
